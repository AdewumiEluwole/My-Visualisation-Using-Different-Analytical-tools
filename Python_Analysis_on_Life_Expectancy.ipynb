```python
PATH='/content/Adewumi_Eluwole_Healthcare_Dataset.csv'
```

```python
Healthcare_data=pd.read_csv(PATH)
```

```python
#This healthcare dataset categorised the countries into by status(developing or developed). 
#Retrieving the developing countries in the dataset
Developing_countries=Healthcare_data.loc[Healthcare_data['Status']== 'Developing']
Developing_countries
```

```python
#Retieving the developed countries from the dataset
Developed_countries=Healthcare_data.loc[Healthcare_data['Status']== 'Developed']
Developed_countries
```

```python
#The correlation between life expectancy and the factors that affect it in the developing countries
plt.figure(figsize=(15,10))
sns.heatmap(Developing_countries.corr(), annot =True, linewidths = 4)
```

```python
#The correlation between life expectancy and the factors that affect it in the developed countries
plt.figure(figsize=(15,10))
sns.heatmap(Developed_countries.corr(), annot =True, linewidths = 4)
```

```python
#The correlation between life expectancy and the factors that affect it in both developing and developed countries
plt.figure(figsize=(15,10))
sns.heatmap(Healthcare_data.corr(), annot =True, linewidths = 4)
```

```python
#Using lineplot to show the sum of adult mortality in each continent
#The plot shows that Africa has the highest Adult mortality rate
plt.figure(figsize=(14,6))
plt.title('Sum of Adult Mortality in Each Continent')
sns.lineplot(x='Continent', y='Adult Mortality', data=Healthcare_data, estimator=np.sum, label='sum', ci=None)
```

```python
#Showing the sum of Under five deaths in each continent using lineplot
#the plot shows that Asia has the highest under five mortality
plt.figure(figsize=(14,6))
plt.title('Sum of Under five deaths in Each Continent')
sns.lineplot(x='Continent', y='Under-five deaths', data=Healthcare_data, estimator=np.sum, label='sum', ci=None)
```

```python
#Checking the mean of Adult Mortality in the dataset in respect to years
#Explaination: 2004 has the highest Adult mortality rate. Generally, the rate has reduced drastically over the years.
plt.figure(figsize=(14,6))
plt.title('Mean of Adult Mortality over the Years')
sns.lineplot(x='Year', y='Adult Mortality', data=Healthcare_data, estimator=np.mean, label='sum')
```

```python
#Using the line plot to show the maximum life expectancy in the developed and the developing countries in respect to year
plt.figure(figsize=(14,6))
plt.title('The Maximum Life Expectancy in developed and developing countries')
sns.lineplot(x='Year', y='Life Expectancy', data=Healthcare_data, hue='Status', estimator=np.max, ci=None)
plt.legend()
```

```python
#Bar chart comparing the total expenditure on health in the developing and the developed countries over the years
plt.figure(figsize=(14,6))
plt.title('Sum of Expenditure on Health in Each Continent')
sns.barplot(x='Year', y='Expenditure_on_Health(GDP)', data=Healthcare_data, estimator=np.sum, hue='Status')
```

```python
plt.figure(figsize=(14,6))
plt.title('Sum of Adult Mortality in developing and developed Countries per year')
sns.barplot(x='Year', y='Adult Mortality', data=Healthcare_data, estimator=np.sum, hue='Status')
```

```python
plt.figure(figsize=(14,6))
plt.title('Sum of Under five deaths in developing and developed Countries per year')
sns.barplot(x='Year', y='Under-five deaths', data=Healthcare_data, estimator=np.sum, hue='Status')
```

```python
#Comparing the immunization intake in the developing and the developed countries
plt.figure(figsize=(14,6))
plt.title('Sum of Polio immunization per year')
sns.barplot(x='Year', y='Polio_immunization', data=Healthcare_data, estimator=sum, hue='Status')
plt.xlabel("Year")
plt.legend()
```
